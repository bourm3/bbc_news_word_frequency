```python
# import libraries for this project
import requests
from bs4 import BeautifulSoup
import re
from nltk.corpus import stopwords
```

```python
# locate the website
url = 'https://www.bbc.co.uk/news/uk-55405299'
```

```python
# get the contents of the website into the variable soup
r = requests.get(url).text
soup = BeautifulSoup(r)
```

```python
#get all the text from the main article on the website
article = soup.find_all("div",attrs={'data-component':"text-block"})
```

```python
#get all the text and clean the data
text = ''
for line in article:
    text += line.text + ' '

text = re.sub(r'\'','',text)
text = re.sub(r'Ã©','e',text)
text = re.sub(r'["-"]',' ',text)
text = re.sub(r'[,.] ',' ',text)
text = re.sub(r'[  ]+',' ',text)
```

```python
# create a list of all the words
words =text.split(' ')
```

```python
# get list of all the stop words to be removed
stopper = stopwords.words('english')
```

```python
# calcualte the frequency of all the words
word_freq= {}
for word in words:
    if word not in stopper:
        word = word.lower()
        if(word not in word_freq):
            word_freq[word] = 1
        else:
            word_freq[word] +=1
```

```python
# find the maximum number of accurancy of a the most common word
max_freq =max(word_freq.values())
```

```python
# divide all frequencies by this max frequency
for word in word_freq.keys():
    word_freq[word] = word_freq[word]/max_freq
```

```python
#sort in desending order
sorted(word_freq.items(),key= lambda x : x[1], reverse=True)
```
